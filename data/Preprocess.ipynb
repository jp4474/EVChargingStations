```python
#get longtitude/latitude for mapping
from tqdm import tqdm
import googlemaps

gmaps_key = key = "mygooglekey"
gmaps = googlemaps.Client(key = gmaps_key)

#empty lists
place_lat = []
place_lng = []


max_lat = 38.0
min_lat = 33.0
max_lng = 132.0
min_lng = 126.0

for place in tqdm(df['Address_Korean']):
    tmp = gmaps.geocode(place, language = "ko")
    

    if tmp:
        tmp_loc = tmp[0].get("geometry")
        tmp_lat = tmp_loc["location"]["lat"]
        tmp_lng = tmp_loc["location"]["lng"]

        if(tmp_lat > max_lat or tmp_lat < min_lat or tmp_lng > max_lng or tmp_lng < min_lng):
            place_lat.append("0")
            place_lng.append("0")

        else:
            place_lat.append(tmp_lat)  
            place_lng.append(tmp_lng)  

    else:
        place_lat.append("0")
        place_lng.append("0")

df['latitude'] = place_lat
df['longitude'] = place_lng
```
