```python
import pandas as pd
```

```python
#read numbers of ev cars in seoul
raw = pd.read_csv("seoul_ev.csv", encoding='cp949')
raw.head()
```

```python
#filter only electric car
raw = raw.loc[raw['연료'] == '전기', ['기준년월', '사용본거지시읍면동_행정동기준', '차명', '연료']]
raw = raw.rename(columns={"사용본거지시읍면동_행정동기준":"주소"})
raw = raw.dropna()
#get district col
raw["구"] = raw["주소"].map(lambda x : x.split()[1])
```

```python
#get total ev cars in each district in seoul
df = raw.groupby("구")["차명"].count().sort_values(ascending=False).to_frame()
df = df.rename(columns={"차명" : "전기차수"})
df.reset_index(inplace = True)
df = df.rename(columns = {'구':'District',	'전기차수':'Count'})
df
```

```python
#read numbers of population in seoul
df_ppl = pd.read_csv("seoul_ppl.csv")
df_ppl.head()
```

```python
# Make the first row the new header
df_ppl.columns = df_ppl.iloc[0]
#get the list of districts in seoul
list_district = list(df['District'])
#filter by districts in seoul
df_ppl = df_ppl[df_ppl['행정구역별(읍면동)'].isin(list_district)]
df_ppl.reset_index(drop=True, inplace = True)
```

```python
# Drop the last 6 rows
df_ppl = df_ppl.iloc[:-6]
#subset cols
df_ppl = df_ppl.iloc[:, :2]
#reset_index
df_ppl.reset_index(drop=True, inplace = True)
#rename cols
df_ppl = df_ppl.rename(columns = {'행정구역별(읍면동)':'District', '총인구 (명)':'Population'})
#make population in int
df_ppl['Population'] = df_ppl['Population'].astype(int)
df_ppl
```

```python
# Merge the DataFrames based on the common column 'District'
fin_df = pd.merge(df, df_ppl, on='District', how='inner')
fin_df
```

```python
district_mapping = {
    '강남구': 'Gangnam-gu',
    '강동구': 'Gangdong-gu',
    '강서구': 'Gangseo-gu',
    '관악구': 'Gwanak-gu',
    '광진구': 'Gwangjin-gu',
    '구로구': 'Guro-gu',
    '노원구': 'Nowon-gu',
    '동대문구': 'Dongdaemun-gu',
    '동작구': 'Dongjak-gu',
    '마포구': 'Mapo-gu',
    '서대문구': 'Seodaemun-gu',
    '서초구': 'Seocho-gu',
    '송파구': 'Songpa-gu',
    '영등포구': 'Yeongdeungpo-gu',
    '용산구': 'Yongsan-gu',
    '은평구': 'Eunpyeong-gu',
    '종로구': 'Jongno-gu',
    '중구': 'Jung-gu',
    '중랑구': 'Jungnang-gu',
    '금천구': 'Geumcheon-gu',
    '도봉구': 'Dobong-gu',
    '강북구': 'Gangbuk-gu',
    '성북구': 'Seongbuk-gu',
    '양천구': 'Yangcheon-gu',
    '성동구': 'Seongdong-gu'
}
fin_df['District'] = fin_df['District'].map(district_mapping)
fin_df.head()
```

```python
#save in csv file
fin_df.to_csv('ev_info_seoul.csv', index=False)
```
